# Поиск лучшей модели
## 1. LogisticRegression

```python
%time
features = df_model.drop(["at_fault","collision_date"], axis=1)
target = df_model["at_fault"]

num_features = ['distance','party_count']
cat_features = ['intersection', 'weather_1', 'primary_collision_factor',
                'road_surface', 'lighting', 'party_sobriety',
                'collision_day_of_week', 'collision_month']

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", StandardScaler())
])

cat_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_features),
        ("cat", cat_transformer, cat_features)
    ]
)

clf_log_regress = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", LogisticRegression(random_state=12345))
])

train_features, test_features, train_target, test_target = train_test_split(
    features, target, test_size=0.2, random_state=12345
)


clf_log_regress.fit(train_features, train_target)

prediction = clf_log_regress.predict(test_features)
prediction_proba = clf_log_regress.predict_proba(test_features)[:,1]

prediction_proba_train = clf_log_regress.predict_proba(train_features)[:,1]
roc_auc_train = roc_auc_score(train_target, prediction_proba_train)

f1_logress = f1_score(test_target, prediction)
roc_auc_logress = roc_auc_score(test_target, prediction_proba)

recall = recall_score(test_target, prediction)
precision = precision_score(test_target, prediction)

print(f"F1: {f1_logress}")
print(f"ROC-AUC:{roc_auc_logress}")
print(f"ROC-AUC_train:{roc_auc_train}")

print(f"recall: {recall}")
print(f"precision: {precision}")
```

## 2.RandomForest

```python
%%time
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="mean")),
])

cat_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_features),
        ("cat", cat_transformer, cat_features)
    ]
)

clf_rand_forest = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(random_state=12345))
])

params_forest = {
    "model__max_depth": np.logspace(0,1.5,7).astype(int),
    "model__n_estimators": np.logspace(1,2,7).astype(int) 
}

grid_forest = GridSearchCV(clf_rand_forest, params_forest, cv = 3, scoring="roc_auc",n_jobs=-1)
grid_forest.fit(train_features, train_target)

best_model_forest = grid_forest.best_estimator_

prediction_forest = best_model_forest.predict(test_features)
prediction_proba_forest = best_model_forest.predict_proba(test_features)[:,1]

roc_auc_forest = roc_auc_score(test_target, prediction_proba_forest)
f1_forest = f1_score(test_target, prediction_forest) 
print(f"F1: {f1_forest}")
print(f"roc_auc: {roc_auc_forest}")

print("\nClassification report:")
print(classification_report(test_target, prediction_forest))
```

## 3. Catboost

```python
%%time
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", StandardScaler())
])

cat_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1))
])


preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_features),
        ("cat", cat_transformer, cat_features)
    ]
)

clf_catboost = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", CatBoostClassifier(
        verbose=0,
        random_state=12345,
    ))
])


params_catboost = {
    "model__learning_rate": np.logspace(-2, -0.5, 7),          # от 0.001 до 0.1
    "model__iterations": range(150,200,5),
    "model__depth": [4,6,8]
}

grid_catboost = GridSearchCV(clf_catboost, params_catboost, cv=3, scoring="roc_auc", n_jobs=-1)
grid_catboost.fit(train_features, train_target)

best_model_catboost = grid_catboost.best_estimator_

prediction_catboost = best_model_catboost.predict(test_features)
predict_proba_catboost = best_model_catboost.predict_proba(test_features)[:, 1]

roc_auc_catboost = roc_auc_score(test_target, predict_proba_catboost)
f1_catboost = f1_score(test_target, prediction_catboost)


print("Лучшие параметры:", grid_catboost.best_params_)
print("ROC-AUC:", round(roc_auc_catboost, 3))
print("F1-score:", round(f1_catboost, 3))
print("\nОтчёт:\n", classification_report(test_target, prediction_catboost))
```

Таким образом, для снижения числа аварий, связанных с алкогольным фактором, можно рассматривать ужесточение санкций: увеличение штрафов или введение более длительных сроков лишения прав. Остальные типы погоды (туман, снег, сильный ветер и пр.) встречаются значительно реже. большое абсолютное число ДТП в ясную погоду объясняется не столько её опасностью, сколько высокой частотой встречаемости таких условий. Можно сделать вывод о наличии более широкого спектра причин, связанных с увеличением риска на дорогах. Таким образом, хотя погодные условия являеются одним из факторов, увеличивающих вероятность ДТП, данные подчеркивают важность рассмотрения и учета других аспектов безопасности на дорогах (например как: состояние опьянения)

## Вывод

В рамках проекта была проведена оценка возможности создания системы прогнозирования риска ДТП при выборе маршрута движения на основе исторических данных.
Для решения задачи классификации были протестированы несколько моделей: логистическая регрессия, дерево решений, XGBoost, LightGBM и CatBoost.

Лучшие результаты показали модели градиентного бустинга (CatBoost, LightGBM и XGBoost) — они продемонстрировали примерно одинаковое качество по метрикам ROC-AUC и F1-score, что указывает на их способность хорошо улавливать сложные нелинейные зависимости между признаками.
При этом логистическая регрессия и дерево решений также показали неплохие результаты, что говорит о том, что данные содержат достаточно чёткие закономерности, связанные с вероятностью ДТП.

Наиболее значимыми признаками, влияющими на вероятность ДТП, оказались:
1. party_sobriety — состояние водителя (трезвый или в состоянии опьянения);
2. weather_1 — погодные условия (ясно, облачно, дождь и т.д.).

Точность модели напрямую зависит от качества и полноты данных, поэтому важно регулярно обновлять и очищать их.В будущем можно улучшить модель, добавив больше контекстных признаков — например, 
плотности трафика, дорожные работы, информацию о водителе (возраст, стаж, история ДТП);
